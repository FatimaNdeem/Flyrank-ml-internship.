## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane is **Refresh / Content Opportunity Scoring**, which is a **scoring** task. The goal is to assign each webpage a score indicating how strongly it should be considered for a content refresh. Pages with higher scores can then be prioritized by the SEO or content team for review and updating.

```python
print("ML Task Type: Scoring")
print("Dataset Shape:", df.shape)
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is a predicted content opportunity score that estimates how likely a webpage is to benefit from a content refresh. Since the dataset does not contain a direct label indicating whether a page should be refreshed, the target would be created as a proxy using observed features such as impressions, clicks, search volume, trend direction, and trend percentage.

```python
df.columns
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

A suitable success metric is **Mean Absolute Error (MAE)** because the task predicts a continuous score. A lower MAE means the predicted scores are closer to the expected values, making the results more useful for prioritizing webpages for content refresh.

```python
print("Chosen Metric: Mean Absolute Error (MAE)")
df[["trend_pct"]].describe()
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

The unit of analysis is **one webpage**. Each row in the dataframe represents one anonymized webpage together with its search performance and content-related features.

```python
print("Dataset Shape:", df.shape)
df.head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule such as "refresh pages with low traffic" would ignore many important factors that influence webpage performance. Features such as impressions, clicks, search volume, and other content signals interact in complex ways. Machine learning can learn these patterns from historical data and help rank or score webpages for refresh, providing better decision support than a fixed if-else rule.

```python
print("Machine learning combines multiple signals to make better recommendations than a single fixed rule.")
```
